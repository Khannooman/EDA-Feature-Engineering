```python
df1 = titanic.groupby(["sex" , "alive"])["pclass"].count().reset_index()
df1
```

```python
df2 = titanic.groupby(["sex"])["pclass"].count().reset_index()
df2
```

```python
# percentages of survival of M/F
df3 = pd.merge(left = df1 , right = df2 , on = "sex")
df3["Survival_percentage"] =( df3["pclass_x"]/df3["pclass_y"])*100
df3
df4 = df3[df3["alive"] == "yes"]
df4
sns.barplot(data = df4 , y = "Survival_percentage" , x = "sex" 
        )
 
plt.show()
```
